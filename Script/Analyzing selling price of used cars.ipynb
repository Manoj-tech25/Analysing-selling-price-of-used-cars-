## Import and read

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
dataset = pd.read_csv('archive/output.csv')
```

```python
dataset.head()
```

```python
dataset=dataset.iloc[:,1:]
```

```python
dataset
```

```python
headers = ["symboling", "normalized-losses", "make", 
           "fuel-type", "aspiration","num-of-doors",
           "body-style","drive-wheels", "engine-location",
           "wheel-base","length", "width","height", "curb-weight",
           "engine-type","num-of-cylinders", "engine-size", 
           "fuel-system","bore","stroke", "compression-ratio",
           "horsepower", "peak-rpm","city-mpg","highway-mpg","price"]
dataset.columns=headers
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.isna().sum()
```

```python
dataset['city-mpg'] = 256/dataset['city-mpg']
dataset.rename(columns = {'city_mpg': "city-L / 100km"},inplace=True)

print(dataset.columns)
dataset.dtypes
```

```python
dataset.price.unique()
```

```python
dataset=dataset[dataset.price != '?']
```

```python
dataset['price']=dataset['price'].astype(int)
dataset.dtypes
```

```python
dataset['length']=dataset['length']/dataset['length'].max()
dataset['height']=dataset['height']/dataset['height'].max()
dataset['width']=dataset['width']/dataset['width'].max()
# binning- grouping values
bins = np.linspace(min(dataset['price']), max(dataset['price']), 4) 
group_names = ['Low', 'Medium', 'High']
dataset['price-binned'] = pd.cut(dataset['price'], bins, 
                              labels = group_names, 
                              include_lowest = True)
print(dataset['price-binned'])
plt.hist(dataset['price-binned'])
plt.show()
```

```python
# categorical to numerical variables
pd.get_dummies(dataset['fuel-type']).head()

# descriptive analysis
# NaN are skipped
dataset.describe()
```

```python
# examples of box plot
plt.boxplot(dataset['price'])
```

```python

# by using seaborn
sns.boxplot(x ='drive-wheels', y ='price', data = dataset)

# Predicting price based on engine size
# Known on x and predictable on y
plt.scatter(dataset['engine-size'], dataset['price'])
plt.title('Scatterplot of Enginesize vs Price')
plt.xlabel('Engine size')
plt.ylabel('Price')
plt.grid()
plt.show()
```

```python
# Grouping Data
test = dataset[['drive-wheels', 'body-style', 'price']]
data_grp = test.groupby(['drive-wheels', 'body-style'], 
                         as_index = False).mean()

data_grp
```

```python
# pivot method
data_pivot = data_grp.pivot(index = 'drive-wheels',
                            columns = 'body-style')
data_pivot

# heatmap for visualizing data
plt.pcolor(data_pivot, cmap ='RdBu')
plt.colorbar()
plt.show()
```

```python
# Analysis of Variance- ANOVA
# returns f-test and p-value
# f-test = variance between sample group means divided by 
# variation within sample group
# p-value = confidence degree
data_annova = dataset[['make', 'price']]
grouped_annova = data_annova.groupby(['make'])
annova_results_l = sp.stats.f_oneway(
                             grouped_annova.get_group('honda')['price'],
                             grouped_annova.get_group('subaru')['price']
                                    )
print(annova_results_l)

# strong corealtion between a categorical variable
# if annova test gives large f-test and small p-value

# Correlation- measures dependency, not causation
sns.regplot(x ='engine-size', y ='price', data = dataset)
plt.ylim(0, )
```
